```python
import streamlit as st
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data["Order Date"] = pd.to_datetime(
    data["Order Date"],
    dayfirst=True
)
```

```python

monthly_sales = (
    data.groupby(
        pd.Grouper(
            key="Order Date",
            freq="ME"
        )
    )["Sales"]
    .sum()
)
```

```python
monthly_sales.head(5)
```

```python
from prophet import Prophet
```

```python
#preparing data for forecasting

forecast_data = monthly_sales.reset_index()

forecast_data.columns = ["ds", "y"]

forecast_data.head()
```

```python

model = Prophet()

model.fit(forecast_data)
```

```python
future = model.make_future_dataframe(
    periods=6,
    freq="ME"
)
```

```python
#predict future sales
#actual prediction
forecast = model.predict(future)
```

```python
forecast[["ds","yhat"]].tail(10)
#viewing of th e prediions
```

```python
#saving forecast time series results

forecast_results = forecast[["ds","yhat"]]

forecast_results.to_csv(
    "sales_forecast.csv",
    index=False
)
```
